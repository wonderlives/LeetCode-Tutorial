```python
## 766. Toeplitz Matrix
### Question: https://leetcode.com/problems/toeplitz-matrix/description/
# Solution:
class Solution(object):
    def isToeplitzMatrix(self, matrix):
        """
        :type matrix: List[List[int]]
        :rtype: bool
        """
        # the matrix is given as M * N
        m = len(matrix)
        n = len(matrix[0])
        
        # the starting point is 0,0
        i = 0 
        j = 0
        
        # Work by row first
        while i < m:
            point = (i,j)
            if not self.isAllMatch(matrix, point):
                return False 
            i += 1
        
        # Only need to work by column in the first row
        while j < n:
            point = (i,j)
            if not self.isAllMatch(matrix, point):
                return False
            j += 1
        
        # If no violation, return True
        return True
        
    def isAllMatch (self, matrix, point):
        m = len(matrix)
        n = len(matrix[0])
        i = point[0]
        j = point[1]
        # Be very careful about the index
        while i < (m) and j < (n):
            if i == m-1 or j == n-1:
                return True
            if  matrix[i][j] == matrix[i+1][j+1]:         
                i += 1
                j += 1
            else:
                return False
        return True
        
# testcase: 
# m = [[1,2,3,4],[5,1,2,3],[9,5,1,2]]
# solution = Solution()
# solution.isToeplitzMatrix(m)
```

```python
# 1. Two Sum
## https://leetcode.com/problems/two-sum/description/

# The key to this question is to use hashmap, here in Python is the dictionary
# As if use 2 loops, it is O(n)
# Use hashmap, here we only need to run it once. 
class Solution(object):
    def twoSum(self, nums, target):
        """
        :type nums: List[int]
        :type target: int
        :rtype: List[int]
        """
        storeDict = {}
        for count, elem in enumerate(nums):
            test = target - elem
            if test in storeDict:
                return [storeDict[test], count]
            else:
                storeDict[elem] = count
```

```python
# 400. Nth Digit
# https://leetcode.com/problems/nth-digit/description/
# the key idea is about range selection. a brutal force is O(n)
# here it can be O(nlogn)
# another important point is about the INDEX is
# different from the number itself!!
class Solution(object):
    def findNthDigit(self, n):
        """
        :type n: int
        :rtype: int
        """
        digits = 1
        currTotal = 0
        start = 1
        i = 1
        base = 9
        while True: 
            currTotal = currTotal + base * digits
            if currTotal > n:
                trim = currTotal - base * digits
                n = n - trim
                x = digits - 1
                if x == 0:
                    start = 0
                else:
                    start = int(str('9'*x))
                break
            base *= 10
            digits += 1 
            i += 1      
        jump = n//digits 
        left = n%digits       
        start += jump
        if left == 0:
            return int(str(start)[-1])
        else:
            return int(str(start+1)[left-1]
```

```python
#97. Maximum Depth of Binary Tree 
# the chanllenge in this problem is the recursive way to find the max depth.
# notice the ending condition is set at when root is None, return 0
# then we can recursively call the function, to avoid store ALL the depth we have calculated, 
# we can just use max(left, right) to get the max depth 

class Solution(object):
    def maxDepth(self, root):
        # This is the ending condition
        if root is None:
            return 0
        
        # this is a compact way to write the solution
        return max(self.maxDepth(root.left), self.maxDepth(root.right))+1
    
    def maxDepth1(self, root):
    # write your code here
        if root is None:
            return 0
        left = self.maxDepth1(root.left)
        right = self.maxDepth1(root.right)
        return max(left,right) + 1

    def maxDepth(self, root):
        self.depth = 0
        self.helper(root, 1)
        return self.depth
        
    def helper(self, root, curr):
        if root is None:
            return self.depth
        
        if curr > self.depth:
            self.depth = curr
        
        self.helper(root.left, curr+1)
        self.helper(root.right, curr+1)
```

```python
# 597. Subtree with Maximum Average
# http://www.lintcode.com/en/problem/subtree-with-maximum-average/
class Solution:
    """
    @param: root: the root of binary tree
    @return: the root of the maximum average of subtree
    """
    def findSubtree2(self, root):
        # write your code here
        self.node = None
        self.average = -float('inf')
        self.helper(root)
        return self.node
        
    def helper(self, root):
        
        if root is None:
            return 0,0
        
        leftSize, leftSum = self.helper(root.left)
        rightSize, rightSum = self.helper(root.right)
        
        treeSize = leftSize + rightSize + 1
        treeSum = leftSum + rightSum + root.val
        
        average = treeSum * 1.0 / treeSize
        
        if average > self.average:
            self.average = average
            self.node = root
    
    
        return treeSize, treeSum
```

```python
## 7. Reverse Integer
## https://leetcode.com/problems/reverse-integer/description/
class Solution(object):
    def reverse(self, x):
        """
        :type x: int
        :rtype: int
        """
        temp = str(x)
        if temp[0] == "-":
            result = -int(temp[:0:-1])
        else:
            result = int(temp[::-1])
            
        if abs(result) < 2**31 -1:
            return result
        else:
            return 0
        
# A more concise approach:
def reverse(self, x):
    s = (x>0)-(x<0)
    r = int('s*x'[::-1])
    return s*r * (r < 2**31)
# The tricky part for this question is the assess the +/-  and make sure it is between -2*31 - 2*31
# also, notice that cmp(x,y) is no longer available
# replace with (a>b) - (a<b) 
# this is kind funny... >-<... hahahaha
```

```python
## 9. Palindrome Number
## https://leetcode.com/problems/palindrome-number/description/
class Solution(object):
    def isPalindrome(self, x):
        """
        :type x: int
        :rtype: bool
        """
        if x < 0:
            return False
        
        ranger = 1 
        
        while x/ranger >= 10:
            ranger *= 10
            
        while x:
            left = x / ranger
            right = x % 10
            if left != right:
                return False
            x = (x%ranger)/10
            ranger = ranger / 100
        
        return True
# The hard part for this question is to find left and right without extra O(n) memory
# Remember: for XXX to get ranger = 100: while x/ranger >= 10: ranger *= 10
# Also, negative number cannot be panlindrome numbers.
# Reverse string works, but nums reverse may overflow, pay attention!
```

```python
## 771. Jewels and Stones
## https://leetcode.com/problems/jewels-and-stones/description/
class Solution(object):
    def numJewelsInStones(self, J, S):
        """
        :type J: str
        :type S: str
        :rtype: int
        """
        stones = list(S)
        count = 0 
        for stone in stones:
            if stone in J:
                count += 1
        return count
# Comprehension
def numJewelsInStones(self, J, S):
    return sum(s in J for s in S)
def numJewelsInStones(self, J, S):
    return sum(map(S.count, J))   
def numJewelsInStones(self, J, S):
    return sum(map(J.count, S))

# An easy question, but need to pay attention to the conprehension.
```
